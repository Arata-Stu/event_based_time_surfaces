```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

## Generate image frame sequence

```python
radius = 30
angle = 5
x0 = 23
y0 = 34

ims = []

for i in range(10):
    im = np.zeros((100, 100), dtype=np.uint8)
    x1 = x0 + radius * np.cos(np.deg2rad(angle))
    y1 = y0 + radius * np.sin(np.deg2rad(angle))
    cv2.line(im, (y0, x0), (int(y1), int(x1)), (255, 0, 0), 2)
    
    ims.append(im)

    if True:
        plt.imshow(im)
        plt.show()

    angle += 4
```
